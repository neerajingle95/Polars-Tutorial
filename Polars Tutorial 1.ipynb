```python
import polars as pl
```

```python
pl.LazyFrame({
    "name": ["Batman", "Gandalf"],
    "age": [35, 55000]
})
```

```python
type(pl.LazyFrame({
    "name": ["Batman", "Gandalf"],
    "age": [35, 55000]
}))
```

```python
pl.LazyFrame({
    "name": ["Batman", "Gandalf"],
    "age": [35, 55000]
}).collect()
```

```python
type(pl.LazyFrame({
    "name": ["Batman", "Gandalf"],
    "age": [35, 55000]
}).collect())
```

```python
# Characters with age > 100
pl.LazyFrame({
    "name": ["Batman", "Gandalf"],
    "age": [35, 55000]
}).filter(pl.col("age") > 100).collect()
```

```python
# Total age of the characters
pl.LazyFrame({
    "name": ["Batman", "Gandalf"],
    "age": [35, 55000]
}).select(pl.col("age").sum()).collect()
```
